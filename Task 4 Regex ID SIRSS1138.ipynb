```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#df = pd.read_csv('case_time_series.csv')

Y = df.iloc[61:,1].values
R = df.iloc[61:,3].values
D = df.iloc[61:,5].values
X = df.iloc[61:,0]

plt.figure(figsize=(30,8)) #To have control over the aesthetics of the graph such as labels, titles,...
#color and size we shall apply more functions as shown below.

ax = plt.axes() #This creates a canvas for the graph where the first value 30 is the width argument position and 
#8 is the height argument position of the graph.

ax.grid(linewidth=0.4, color='#8f8f8f')  #.grid function lets you create grid lines across the graph. The width of the
#grid lines can be adjusted by simply passing an argument linewidth and changing its color by passing color argument.

ax.set_facecolor("black") # .set_facecolor lets you set the background color of the graph which over here is black. 
#.set_xlabel and .set_ylabel lets you set the label along both axes whose size and color can be altered .

ax.set_xlabel('\nCountries',size=30,color='#4bb4f2')
ax.set_ylabel('Number of Confirmed Cases per million\n',
			size=25,color='#4bb4f2')


#Still as we can see the dates are overlapping and the labels along axes are not clear enough

#.xticks and .yticks lets you alter the Countries and Confirmed Cases. For the Cases to not overlap with each other we 
#are going to represent them vertically by passing rotation = vertical. To make the ticks easily readable we change the font
#color to Red and size to 10.
plt.xticks(rotation='vertical',size='10',color='Red')
plt.yticks(size=20,color='Red')
plt.tick_params(size=10,color='white') #.tick_params lets you alter the size and color of the dashes which look act like a 
#bridge between the ticks and graph.

for i,j in zip(X,Y):
	ax.annotate(str(j),xy=(i,j+100),color='white',size='15')
    
#.annotate lets you "add notes" on the graph. code to annotate the plotted points by running a for loop which . 
#plots at the plotted points. The str(j) holds the Y variable which is Confirmed cases per million Any string passed will be plotted. The XY 
#is the coordinates where the string should be plotted. And finally the color and size can be defined
	
ax.annotate('Second Lockdown 15th April',
			xy=(15.2, 860),
			xytext=(19.9,500),
			color='Yellow',
			size='33',
			arrowprops=dict(color='Red',
							linewidth=0.025))
#To annotate an arrow pointing at a position in graph and its tail holding the string we can define arrowprops argument along with its 
#tail coordinates defined by xytext. Note that arrowprops alteration can be done using a dictionary

plt.title("COVID-19 IN :  Confrimed Cases\n",
		size=50,color='#28a9ff')

ax.plot(X,Y,
		color='#1F77B4',
		marker='o',
		linewidth=4,
		markersize=12,
		markeredgecolor='#035E9B')
```

```python
slices = [16480485, 654036, 9468087,6358362]  #total values of all coumns and then plotting a piechart on that basis
activities = ['Confirmed', 'Death', 'Recorded','Active']

cols=['#4C8BE2','#00e061','#fe073a','#ffff00']
exp = [0.1,0.01,0.01,0.1] # fr spacing b/w the slices 

plt.pie(slices,labels=activities,
		textprops=dict(size=25,color='black'),
		radius=2,
		colors=cols,
		autopct='%2.2f%%',
		explode=exp,
		shadow=True,
		startangle=90)

plt.title('The Covid Pie\n\n',color='#4fb4f2',size=40)
```
